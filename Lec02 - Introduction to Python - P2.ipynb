#### $\color{green}{\text{Example 1}}$
1. Create a variable that stores the age of five (5) people
2. Create a variable that stores that stores three (3) fruit names
3. Create a variable that stores three (3) eye colors

```python
#Variables that store the age of five people
age = [56,36,85,96,40,75]
print (age)

#Create a variable that stores three fruit names
fruits = ["orange","apple","Pair"]
print (fruits)

#Create a variable that stores three eye colors
eyeColors = ["Blue", "Green", "Brown"]
print(eyeColors)
```
